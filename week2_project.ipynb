# Remove Duplicates

```python
df = df.drop_duplicates()
```

# Fill Missing Age Values

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

# Fill Missing Embarked Values

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

# Check Missing Values Again

```python
df.isnull().sum()
```

## 7. Key Insights from Data Analysis

🧠 **Summary of Findings:**

1. **Gender Impact (STRONGEST FACTOR)** - Females had a higher survival rate than males (women and children first protocol)

2. **Passenger Class Impact** - 1st class passengers had better survival chances compared to 2nd and 3rd class passengers due to better access to lifeboats

3. **Age Impact** - Younger passengers had slightly higher survival probability (children were prioritized)

4. **Feature Correlations** - Passenger class, gender, and age are the strongest predictors of survival

5. **Overall Survival Rate** - Approximately 38% of passengers survived the disaster

### Conclusions:
- Social class was a major determinant of survival
- Gender was crucial (women and children first)
- Age mattered, especially for younger passengers
- These factors combined heavily influenced who survived the Titanic disaster

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

## 6. Correlation Heatmap (FEATURE RELATIONSHIPS)

**Insight:** Younger passengers had slightly higher survival probability. Children were prioritized during evacuation, contributing to better survival rates for lower ages.

```python
sns.boxplot(x='survived', y='age', data=df)
plt.title("Age vs Survival")
plt.show()
```

## 5. Survival vs Age (Deep Insight)

```python
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()
```

## 4. Age Distribution

**Insight:** 1st class passengers had significantly better survival chances compared to 2nd and 3rd class passengers, likely due to better access to lifeboats.

```python
sns.countplot(x='survived', hue='pclass', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

## 3. Survival by Passenger Class

**Insight:** Females had a significantly higher survival rate than males. This reflects the historical "women and children first" protocol observed during the Titanic disaster.

```python
sns.countplot(x='survived', hue='sex', data=df)
plt.title("Survival by Gender")
plt.show()
```

## 2. Survival by Gender (IMPORTANT INSIGHT)

```python
sns.countplot(x='survived', data=df)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```
